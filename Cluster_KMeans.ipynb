```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate random data points
X, y = make_blobs(n_samples=100, centers=3, n_features=3, random_state=11)

# Fit a KMeans model
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

# Plot the clusters in 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X[:,0], X[:,1], X[:,2], c=model.labels_, s=10)
# Set the plot limits and labels
ax.set_xlim([-10, 10])
ax.set_ylim([-10, 10])
ax.set_zlim([-10, 10])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
# ax.scatter(-10, 0, 10, s=100, c='r', marker='o')
# Set the viewing angle
ax.view_init(elev=30, azim=45)

plt.show()
```

```python
# Predict the cluster label for a new data point
X_new = np.array([[-10, 0, 10]])  # new data point
label = model.predict(X_new)[0]  # predicted cluster label
print("New data point belongs to cluster:", label)
```

```python
import numpy as np
import os

# Set the path to the folder containing the text files
folder_path = "/home/yhsharaf/Desktop/MBZUAI-ML703/TestingWater"

#Empty array
empty_array = []

# Loop over each file in the folder
for filename in os.listdir(folder_path):
    # Check if the file is a text file
    if filename.endswith(".txt"):
        # Open the file in read mode
        with open(os.path.join(folder_path, filename), "r") as f:
            # Read the last three lines of the file
            lines = f.readlines()[-1:]

            # Extract the x, y, and z values from the lines
            for line in lines:
                values = line.strip().split(",")
                x = float(values[0])
                y = float(values[1])
                z = float(values[2])
                # print(f"x: {x}, y: {y}, z: {z}")

            # Store the values in an array and print the array
            empty_array.append([x,y,z])
print(empty_array)
```

```python
X = np.array(empty_array)
min_val = X.min()
max_val = X.max()
normalized_array = (X - min_val) / (max_val - min_val)
print(normalized_array)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

X = normalized_array

# Fit a KMeans model
model = KMeans(n_clusters=3, random_state=42)
model.fit(X)

# Define colors for each cluster
colors = ['gold',
'orange',
'orangered',
'greenyellow',
'turquoise',
'cyan',
'dodgerblue',
'blue',
'blueviolet',
'fuchsia',
'deeppink',
'crimson']

# Plot the clusters in 3D with specific colors
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
for i in range(len(X)):
    ax.scatter(X[i, 0], X[i, 1], X[i, 2], c=colors[model.labels_[i]], s=125,marker='o')
# Set the plot limits and labels
ax.set_xlim([0, 1])
ax.set_ylim([0, 1])
ax.set_zlim([0, 1])
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
# Set the viewing angle
ax.view_init(elev=45, azim=180+45)

plt.show()
```

```python
import numpy as np
import os

# Set the path to the folder containing the text files
folder_path = "/home/yhsharaf/Desktop/MBZUAI-ML703/TestingWater/"

#Empty array
empty_array = []
i = 0
# Loop over each file in the folder
for i in range(len(sort_array)):
    file_path = os.path.join(folder_path, sort_array[i])
    with open(file_path, "r") as f:

        # Read the last three lines of the file
        lines = f.readlines()

        for i, line in enumerate(lines):
            if "XYZ:" in line:
                xyz_index = i
                break
        
        next_line = lines[xyz_index].strip()  
        # get the line after "XYZ:" and remove any leading/trailing whitespace
        next_line = next_line.replace("XYZ: ","").split(",")
        # print(next_line)
        x = float(next_line[0])
        y = float(next_line[1])
        z = float(next_line[2])
        # print(f"x: {x}, y: {y}, z: {z}")

        # Store the values in an array and print the array
        empty_array.append([x,y,z])
        i=1
print(empty_array)
```

```python
X = np.array(empty_array)
min_val = X.min()
max_val = X.max()
normalized_array = (X - min_val) / (max_val - min_val)
print(normalized_array)
```
